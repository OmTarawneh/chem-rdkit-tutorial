# Getting Started with RDKit

A quick intro to creating molecules, displaying them, and computing basic properties.

```python
from rdkit import Chem
from rdkit.Chem import Descriptors
from rdkit.Chem import rdMolDescriptors
```

## Create a molecule from SMILES

[SMILES](https://en.wikipedia.org/wiki/Simplified_Molecular_Input_Line_Entry_System) is a string notation for describing molecular structures. Let's create caffeine:

```python
caffeine = Chem.MolFromSmiles("CN1C=NC2=C1C(=O)N(C(=O)N2C)C")
caffeine
```

## Basic molecular properties

```python
print(f"Formula:         {rdMolDescriptors.CalcMolFormula(caffeine)}")
print(f"Molecular weight: {Descriptors.MolWt(caffeine):.2f}")
print(f"Num atoms:       {caffeine.GetNumAtoms()}")
print(f"Num bonds:       {caffeine.GetNumBonds()}")
```

## Aspirin smile


```python
asp = Chem.MolFromSmiles("CC(=O)OC1=CC=CC=C1C(=O)O")
asp
```

```python
print(f"Formula:         {rdMolDescriptors.CalcMolFormula(asp)}")
print(f"Molecular weight: {Descriptors.MolWt(asp):.2f}")
print(f"Num atoms:       {asp.GetNumAtoms()}")
print(f"Num bonds:       {asp.GetNumBonds()}")
```
